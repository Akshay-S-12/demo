```python
import seaborn as sns
df=sns.load_dataset('tips')
print(df)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.dtypes)
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
#joint plot
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='size',data=df)
```

```python
#pair plot
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
df['day'].value_counts()
```

```python
sns.pairplot(df,hue='time')
```

```python
#count plot
sns.countplot(x='day',data=df,color='r')
```

```python
sns.countplot(x='day',data=df,color='r',hue='smoker')
```

```python
sns.countplot(x='day',data=df,color='r',hue='sex')
```

```python
#barplot
sns.barplot(x='total_bill',y='day',data=df)
```
