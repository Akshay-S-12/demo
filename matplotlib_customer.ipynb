```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/customer1.csv')
print(df)
```

```python
print(df.isna().sum())
```

```python
df1=df.fillna('india')
print(df1)
```

```python
df.dtypes
```

```python
x=df['loc'].mode()[0]
print(x)
df['loc'].fillna(x,inplace=True)
print(df)
```

```python
df1=df['loc'].value_counts()
print(df1)
x1=df1.index
y1=df1.values
plt.bar(x1,y1,color='r')
plt.xlabel('location')
plt.ylabel('count')
plt.title('location count')
```

```python
y=df1.values
lab=df1.index
plt.pie(y,labels=lab,autopct='%.1f%%')
plt.legend(loc='upper left')
plt.tight_layout()
plt.title('location count')
```

```python
df2=df['prof'].value_counts()
print(df1)
x1=df2.index
y1=df2.values
plt.barh(x1,y1,color='r')
plt.xlabel('prof')
plt.ylabel('count')
plt.title('profession count')
```

```python
y=df2.values
lab=df2.index
plt.pie(y,labels=lab,autopct='%.1f%%')
plt.legend(loc='upper left')
plt.tight_layout()
plt.title('location count')
```
